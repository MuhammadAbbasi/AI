```python
class Deque:
    def __init__(self):
        self._data=[None] * 10
        self._front =- 1
        self._back=0
    def __len__(self):
        return len(self._data)
    def isempty (self):
        return self._data == 0
    def first (self):
        return self._data[self._back]
    def last (self):
        return self._data[self._front]
    def add_first(self,num):
        self._data[self._front] = num
        self._front -= 1
    def add_last(self,num):
        self._data[self._back] = num
        self._back += 1
    def delete_first(self):
        self._front += 1
        self._data[self._front] = None
    def delete_last(self):
        self._back -= 1
        self._data[self._back] = None
    def printlist(self)    :
        print(self._data)
a=Deque()
for x in range(10):
    a.add_first(x)
    a.printlist()
for y in range(10):
    a.delete_last()
    a.printlist()

for z in range(10):
    a.add_last(z)
    a.printlist()
for v in range(10):
    a.delete_first()
    a.printlist()
```
